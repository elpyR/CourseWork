```python
class Customer:
    def __init__(self, name, email, phone):
        self.name = name
        self.email = email
        self.phone = phone
        self.orders = []

    def place_order(self, order):
        self.orders.append(order)

    def update_contact_details(self, email, phone):
        self.email = email
        self.phone = phone

class Item:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price
        self.category = None

    def set_category(self, category):
        self.category = category

class Category:
    def __init__(self, name):
        self.name = name
        self.items = []

    def add_item(self, item):
        self.items.append(item)

class Order:
    def __init__(self, order_id, date, status):
        self.order_id = order_id
        self.date = date
        self.status = status
        self.customer = None
        self.order_items = []

    def set_customer(self, customer):
        self.customer = customer

    def add_item(self, item, quantity):
        order_item = OrderItem(item, quantity)
        self.order_items.append(order_item)

    def remove_item(self, item):
        self.order_items = [oi for oi in self.order_items if oi.item != item]

    def update_status(self, status):
        self.status = status

class OrderItem:
    def __init__(self, item, quantity):
        self.item = item
        self.quantity = quantity

class Payment:
    def __init__(self, payment_id, payment_date, total_amount):
        self.payment_id = payment_id
        self.payment_date = payment_date
        self.total_amount = total_amount
        self.order = None

    def set_order(self, order):
        self.order = order

# Test Cases
if __name__ == "__main__":
    # Create a customer
    customer = Customer("John Doe", "john@example.com", "123-456-7890")

    # Create categories and items
    category1 = Category("Electronics")
    category2 = Category("Clothing")

    item1 = Item("Laptop", "High-performance laptop", 999.99)
    item2 = Item("Smartphone", "Latest smartphone model", 499.99)
    item3 = Item("T-Shirt", "Comfortable cotton t-shirt", 19.99)

    category1.add_item(item1)
    category1.add_item(item2)
    category2.add_item(item3)

    item1.set_category(category1)
    item2.set_category(category1)
    item3.set_category(category2)

    # Create an order
    order = Order(1, "2023-09-30", "Pending")
    order.set_customer(customer)

    order.add_item(item1, 2)
    order.add_item(item3, 5)

    # Update customer contact details
    customer.update_contact_details("newemail@example.com", "987-654-3210")

    # Update order status
    order.update_status("Shipped")

    # Create a payment
    payment = Payment(12345, "2023-10-01", 1499.95)
    payment.set_order(order)

    # Display order details
    print(f"Order ID: {order.order_id}")
    print(f"Customer: {order.customer.name}")
    print("Ordered Items:")
    for order_item in order.order_items:
        print(f"{order_item.item.name} - Quantity: {order_item.quantity}")
    print(f"Order Status: {order.status}")
    print(f"Payment Amount: ${payment.total_amount}")
```
